```python
import json 
import pandas as pd
import numpy as np
path = '../data/masked_dataset/maksed_data.jsonlf'
df = pd.read_json(path, lines=True)
```

```python
# check if partial_counterfactual_text only contain [MASK] tokens
df["pc_only_masked"] = df["partial_counterfactual_text"].apply(lambda x: all(token == "[MASK]" for token in x.split()))
df["text_length"] = df["text"].apply(lambda x: len(x.split()))
df["masked_length"] = df["partial_counterfactual_text"].apply(lambda x: x.count("[MASK]"))
```

```python
import numpy as np
pd.set_option('display.max_rows', None)

def true_percentage(x):
    return x.sum()/len(x)
# create pivot table with df
df_pivot = df.pivot_table(index=["Dataset","Base_Model","Stereo_Type"], values=["pc_only_masked","Mask_Ratio","text_length","masked_length"], aggfunc={"pc_only_masked":true_percentage,"Mask_Ratio":np.mean,"text_length":np.sum,"masked_length":np.sum})
df_pivot["masked_percentage"] = df_pivot["masked_length"]/df_pivot["text_length"]
df_pivot["stereotype_percentage"] = 1 - df_pivot["masked_percentage"]
df_pivot
```
